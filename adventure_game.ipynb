```python
name = input("what is your name: ")
print("wellcome",name,"to this adventure")

answer=input("you are on a dirt road, it has come to an end and you can go to either left or right.which way would you like to go? ").lower()
if answer == "left":
  answer = input("you come to river, you can walk around it or swim across? Type walk to walk around and swim to swim around: ")
  if answer == "swim":
    print("you swam and eaten by alligator")

  elif answer == "walk":
    print("you walk for many miles,ran out of water and you lose the game")

elif answer == "right":
  answer=input("you are on a brige, it looks woobly, do u want to cross it or go back").lower()
  if answer == "back":
    print("you can go back")

  elif answer == "cross":
    answer=input("you cross the bridge and meet with stranger, do u talk with them (yes/no)? ")
    if answer == "yes":
        print("you talk with stranger and gives gold so you win")

    elif answer == "no":
        print("you ignore the stranger so you lose")

    else:
        print("not a valid option so you lost")
    print("")

else:
  print("not a valid option you lose")

print("thank you for joining",name)
```
