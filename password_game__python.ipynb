```python
master_pwd = input("what is the master password? ")
def view():
  with open("password.txt","r") as f:
    for line in f.readlines():
      data=line.rstrip()
      user,passw = data.split("|")
      print("user :",user,"passw: ",passw)



view()

def add():
  name= input("account name: ")
  pwd= input("password: ")

  with open("password.txt","a") as f:
    f.write(name + "|" + pwd + "\n")

add()


while True:
  mode=input("would you like to add new password or see exciting one (view/add)? or press q to quit")
  if mode == "q":
    break
  if mode == "view":
    view()
  elif mode == "add":
    add()
  else:
    print("invalid mode")
```
